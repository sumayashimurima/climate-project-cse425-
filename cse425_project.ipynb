```python
filename = "global_warming_dataset.csv"
```

```python
def data_loading(filename):
    data = []
    with open(filename, 'r', encoding='utf-8') as File:
        header = File.readline().strip().split(",")  # read header manually
        for line in File:
            parts = line.strip().split(",")
            if len(parts) != len(header):
                continue  # skip malformed rows
            try:
                record = {
                    "Country": parts[0],
                    "Years": int(parts[1]),
                    "Temperature_Anomaly": float(parts[2]),
                    "CO2": float(parts[3]),
                    "GDP": float(parts[4]),
                    "Extreme_Weather_Events": float(parts[5])  # FIX: use float
                }
                data.append(record)
            except ValueError:
                # Skip row if conversion fails
                continue
    return data
```

```python
# --------------------------
# 1. Search by Country
# --------------------------
def search_by_country(data, country_name):
    results = []
    for i in range(len(data)):
        if data[i]["Country"].lower() == country_name.lower():
            results.append(data[i])
    return results
```

```python
# --------------------------
# 2. Calculate Average GDP
# --------------------------
def average_gdp(data, country_name):
    total = 0
    count = 0
    for i in range(len(data)):
        if data[i]["Country"].lower() == country_name.lower():
            total += data[i]["GDP"]
            count += 1
    if count == 0:
        return None
    return total / count
```

```python
# --------------------------
# 3. Calculate Total GDP
# --------------------------
def total_gdp(data, country_name):
    total = 0
    for i in range(len(data)):
        if data[i]["Country"].lower() == country_name.lower():
            total += data[i]["GDP"]
    return total
```

```python
# --------------------------
# 4. Sort by GDP (Bubble Sort)
# --------------------------
def sort_by_gdp(data, year, ascending=True):
    # filter manually
    filtered = []
    for i in range(len(data)):
        if data[i]["Years"] == year:
            filtered.append(data[i])

    n = len(filtered)
    # Bubble Sort by GDP
    for i in range(n):
        for j in range(0, n - i - 1):
            if ascending:
                if filtered[j]["GDP"] > filtered[j+1]["GDP"]:
                    filtered[j], filtered[j+1] = filtered[j+1], filtered[j]
            else:
                if filtered[j]["GDP"] < filtered[j+1]["GDP"]:
                    filtered[j], filtered[j+1] = filtered[j+1], filtered[j]
    return filtered

```

```python
# --------------------------
# 5. Find Top CO2 Emitters (Selection Sort)
# --------------------------
def top_co2_emitters(data, year, top_n=5):
    # filter manually
    filtered = []
    for i in range(len(data)):
        if data[i]["Years"] == year:
            filtered.append(data[i])

    # Selection sort by CO2 (descending)
    n = len(filtered)
    for i in range(n):
        max_idx = i
        for j in range(i+1, n):
            if filtered[j]["CO2"] > filtered[max_idx]["CO2"]:
                max_idx = j
        # swap
        filtered[i], filtered[max_idx] = filtered[max_idx], filtered[i]

    # return top_n emitters
    top_emitters = []
    for i in range(top_n):
        if i < len(filtered):
            top_emitters.append(filtered[i])
    return top_emitters
```

```python
def main():
    filename = "global_warming_dataset.csv"

    start_time = time.time()   # start timer

    data = data_loading(filename)

    print("\n1. Search by Country: Bangladesh")
    bd_data = search_by_country(data, "Bangladesh")
    print("Records Found:", len(bd_data))

    print("\n2. Average GDP of Bangladesh:")
    print(average_gdp(data, "Bangladesh"))

    print("\n3. Total GDP of Bangladesh:")
    print(total_gdp(data, "Bangladesh"))

    print("\n4. Sort by GDP in 2000 (Ascending):")
    sorted_gdp = sort_by_gdp(data, 2000, ascending=True)
    print("Top 5 sorted GDP:", sorted_gdp[:5])

    print("\n5. Top 5 CO2 Emitters in 2000:")
    top_emitters = top_co2_emitters(data, 2000, 5)
    print("Top 5 Emitters:", top_emitters)

    end_time = time.time()     # end timer
    print(f"\n⏱ Runtime: {end_time - start_time:.4f} seconds")


if __name__ == "__main__":
    main()
```
